```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load the image
image = cv2.imread('ressources/mm-binary.png', cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(6, 6))
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Step 2: Define the cross-shaped structuring element (SE)
se = cv2.getStructuringElement(cv2.MORPH_CROSS, (3, 3))
print("Structuring Element (SE):\n", se)
```

```python
# Step 3: Compute morphological gradient
gradient = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, se)

    # Display the gradient image
plt.figure(figsize=(6, 6))
plt.imshow(gradient, cmap='gray')
plt.title('Morphological Gradient')
plt.axis('off')
plt.show()
```
