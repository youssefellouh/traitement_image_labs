```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Dimensions de l'élément structurant
size = 11# Créer un noyau croix (forme cruciforme)
cross_kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (size, size))
# Créer un noyau cercle
circle_kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (size, size))
# Créer un noyau rectangulaire
rect_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (size, size))
# Affichage des noyaux avec Matplotlib
plt.figure(figsize=(10, 4))
plt.subplot(1, 3, 1)
plt.imshow(cross_kernel, cmap='gray')
plt.title('Noyau Croix')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(circle_kernel, cmap='gray')
plt.title('Noyau Cercle')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(rect_kernel, cmap='gray')
plt.title('Noyau Rectangulaire')
plt.axis('off')

plt.tight_layout()
plt.show()
```
